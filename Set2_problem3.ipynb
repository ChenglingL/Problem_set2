```python
# importing module
import numpy as np


# numpy.random.uniform() method
X = np.random.uniform(-1.0, 1.0, 10000)

Y = np.random.uniform(-1.0, 1.0, 10000)

Z1 = X + Y

Z2 = X * Y
```

```python
import matplotlib.pyplot as plt

num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(Z1, num_bins, density=True)

# add a 'best fit' line
y = (-bins[(bins > 0) & (bins < num_bins/2)] + 2)/4
ax.plot(bins[(bins > 0) & (bins < num_bins/2)], y, '--','r')

y = (bins[(bins > -num_bins/2) & (bins < 0)] + 2)/4
ax.plot(bins[(bins > -num_bins/2) & (bins < 0)], y, '--','r')

plt.show()
```

```python
import matplotlib.pyplot as plt

num_bins = 500

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(Z2, num_bins, density=True)

# add a 'best fit' line
y = (-np.log(bins[(bins >= 0) & (bins < num_bins/2)]))/2
ax.plot(bins[(bins >= 0) & (bins < num_bins/2)], y, '--','r')

y = (-np.log(-bins[(bins > -num_bins/2) & (bins < 0)]))/2
ax.plot(bins[(bins > -num_bins/2) & (bins < 0)], y, '--','r')

plt.show()
```
